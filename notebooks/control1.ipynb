```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x),label='carro 1');
ax.plot(x, 2*np.sin(x), label='carro 2');
ax.grid()
ax.set(xlabel='tempo [s]',ylabel='velocidade [m/s]',title='gr√°fico')
ax.legend(loc=0)
```
